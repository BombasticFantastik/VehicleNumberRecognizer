```python
import os
import PIL
from PIL import Image
from torch.utils.data import Dataset
import json
import torch
import matplotlib.pyplot as plt
from torchvision import transforms
```

```python
class NumberDataset(Dataset):
    def __init__(self,path):
        super(NumberDataset,self).__init__()
        img_path=os.path.join(path,'img')
        label_path=os.path.join(path,'ann')
        self.images=[os.path.join(img_path,img) for img in os.listdir(img_path)]
        self.images.sort(reverse=True)
        self.labels=[os.path.join(label_path,label) for label in os.listdir(label_path)]
        self.labels.sort(reverse=True)

        self.trans=transforms.Compose([
            transforms.Resize((64,128)),
            transforms.ToTensor()
        ])

        #проводим зачистку
        
        
        test_labels=[label.split('/')[-1][:-5]) for label in self.labels]
        test_images=[img.split('/')[-1][:-5]) for img in self.images]

        #лейблы
        for idx in range(len(test_labels)):
            if test_labels[idx] not in test_images:
                self.labels.pop(idx)
        #изображения
        for idx in range(len(test_images)):
            if test_images[idx] not in test_labels:
                self.images.pop(idx)
                
        

        # for img in self.images:
        #     print(img.split('/')[-1][:-5])
    def __len__(self):
        return len(self.images)
    def __getitem__(self,idx):
        idx_img=Image.open(self.images[idx])
        idx_label=self.labels[idx]
        with open(idx_label,'r') as file_option:
            jf=json.load(file_option)
            tensor_label=jf['name']

        tensor_img=self.trans(idx_img)

        
        
        return {
            'img':tensor_img,
            'label':tensor_label
        }
```
